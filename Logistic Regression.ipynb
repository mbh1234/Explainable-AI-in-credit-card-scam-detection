```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# Importing metrics
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
# Confusion matrix
confusion = metrics.confusion_matrix(Y_train, X_train_prediction)
print(confusion)
```

```python
TP = confusion[1,1] # true positive 
TN = confusion[0,0] # true negatives
FP = confusion[0,1] # false positives
FN = confusion[1,0] # false negatives
```

```python
# Accuracy
print("Accuracy:-",metrics.accuracy_score(Y_train, X_train_prediction))

# Sensitivity
print("Sensitivity:-",TP / float(TP+FN))

# Specificity
print("Specificity:-", TN / float(TN+FP))

# F1 score
print("F1-Score:-", f1_score(Y_train, X_train_prediction))
```

```python
# classification_report
print(classification_report(Y_train, X_train_prediction))
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
# Confusion matrix
confusion = metrics.confusion_matrix(Y_test, X_test_prediction)
print(confusion)
```

```python
TP = confusion[1,1] # true positive 
TN = confusion[0,0] # true negatives
FP = confusion[0,1] # false positives
FN = confusion[1,0] # false negatives
```

```python
# Accuracy
print("Accuracy:-",metrics.accuracy_score(Y_test, X_test_prediction))

# Sensitivity
print("Sensitivity:-",TP / float(TP+FN))

# Specificity
print("Specificity:-", TN / float(TN+FP))

# F1 score
print("F1-Score:-", f1_score(Y_test, X_test_prediction))
```

```python
# classification_report
print(classification_report(Y_test, X_test_prediction))
```
